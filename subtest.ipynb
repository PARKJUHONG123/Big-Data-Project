```python
sub=subway[['역사명','노선명','역위도','역경도','역사도로명주소']]
```

```python
sub.rename(columns={'역위도':'x','역경도':'y'})
```

```python
headers = {'Authorization' : 'KakaoAK cece2f201ea21342bd507cea8923f5bd'}
k=0
for i in sub['역사도로명주소']:
    body = {'query':i}
    response = requests.request("GET",
                            'https://dapi.kakao.com/v2/local/search/address.json',
                            data=body,
                            headers=headers)
    res = json.loads(response.text)
    if(res['documents']!=[]):
        sub.loc[k,'x']=res['documents'][0]['y']
        sub.loc[k,'y']=res['documents'][0]['x']
    k+=1
```

```python
m=folium.Map(location=[35.8691072,128.5787031],zoom_start=12)
```

```python
sub=sub.fillna(-1)
```
